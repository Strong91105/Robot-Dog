## Load model, perform detection, visualize result

In the first section of this notebook, you will learn how to load and run a YOLO model for detection tasks

````python
# imports packages
from ultralytics import YOLO
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load yolo model
# You can try different models
model = YOLO("yolo11n.pt")

````

````python
# Perform pedection on an image
results = model("https://ultralytics.com/images/bus.jpg")

# Show the results
result = results[0].plot()
rgb_image = cv2.cvtColor(result, cv2.COLOR_BGR2RGB) # Convert color settingss
plt.imshow(rgb_image)
plt.axis('off')
plt.show()
````
